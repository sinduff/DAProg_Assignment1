**Normal Distribution** is more commonly known as the a Bell Curve, but is also called the Gaussian distribution.$^{18,19}$  Wikipedia classifies this distribution as being able to support a 'whole real line'.$^{18}$
<br><br>
In Numpy, the *random.normal* method has 3 parameters$^{18,19}$, namely
1. *loc* - is the mean or the center of the distribution
2. *scale* - is the spread or width (stardard deviation) of the distribution.
3. *size* - this refers to the shape of the returned array.  This is an optional parameter.
<br><br>
The output of the function is an array, which can then be used to create a plot using the Seaborn plotting Library.
<br><br>
The normal distribution function is widely used across many areas.  The key premisis of the normal distribution function is that for a given random variable, the results will converger around the middle or mean of the sample.  Basically, this means that the data will bunch around the middles, with a smaller sample of results at either end.
<br><br>
A real world example of this would be the distrubtion of exam results in a class.  Most of the class would be expected to achieve results that would put them in the middle, with a smaller number of students achieving top or lower grades.
<br><br>
Sample code to demonstrate this is shown below in Examples 1 and 2.
<br><br>
*Example 6* - displays a simple normal distribution curve, with random data and a size of 1000.  A historgram is show to demonstrate the different values making up the curve.
<br><br>
*Example 7* - the code is taken from the NumPy.org page for Normal Distribution.$^{19}$  The code in the first cell sets the loc, scale and size parameters for the object. It also creates an array called 's' and populates with 1000 pieces of random data.
<br><br>
The second block of code creates the histogram and determines how it will look.


```python
#Example 6  
#sample code to demonstrate a simple normal distribution curve for random values

sns.distplot(random.normal(size=1000))

plt.show()
```

```python
#Example 7 - sample code taken from Numpy.org's page on numpy.random (Ref 19)

#set the mean (mu) and standard deviation (signma) for the distrbution
mu, sigma = 0, 0.1 

# create an array 's' and populate with random data, using the normal distribution.
#loc is set to 0
#scale is set to 0.1
#size is set to 1000
s = np.random.default_rng().normal(mu, sigma, 1000)
```

```python
#Create a histogram (using Matplotlib) to display the data, setting the 
#paramets for the graph, including count array 's' values for bins / linecolor etc.
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=3, color='r')
plt.show()
```

**Pareto Distribution** method follows Pareto's Law of an 80 / 20 distribution, i.e. 20% factors cause an 80% outcomes.$^{27,28}$  Wikipedia defines this distribution as a supported on semi-infite intervals, and details that this analysis is generally used to understand financial data.$^{18}$

As outlined above, the Pareto Distribution is generally used to examine / explain financial data.  Practical application of the data have been described as$^{29}$;

<br>
Example 11 below, sourced from NumPy.org, demonstrates how a prateo distribution would work.  Explanations of the code have been included as comments.

```python
#Example 11 - Pareto's Distribution Analysis - 80/20 Rule

# outline parameters (shape) for the default_rng().pareto method 
a = 3.0
#defines the mode of the array, i.e. the number that appears most often
m = 2.0

#defines the array, fills it using the value of a, size of 1000,and multiply by the mode
s = (np.random.default_rng().pareto(a, 1000) + 1) * m
```

```python
#this section of Example 11 defines the graph to be displayed.
# counts the values in the array, and defines the size of the 'bins' of data.
count, bins, _ = plt.hist(s, 100, density=True)
fit = a*m**a / bins**(a+1)
plt.plot(bins, max(count)*fit/max(fit), linewidth=2, color='g')
plt.title("Example 11 - Pareto's Distribution Analysis - 80/20 Rule")
plt.xlabel("Occurence of the number in the array")
plt.ylabel("Number value");
plt.show()
```

### Section 4 - Seeds and Random Number Generation

**Seeds** in NumPy refer to the methods of getting random numbers that are generated by the BitGenerator and Generators in Python.$^{10}$ 
<br><br>
As stated in Section 1, computers produce 'pseudo random' numbers, or num ers that look like they are random, but are not.$^{34}$  A tutorial by Sharp Sight$^{34}$ explains that one of the reasons for this is that computers run on algorithms, therefore, they will produce the same result for a given input.  That is to say, that the output of a computer program cannot be random.
<br><br>
By setting a 'seed' a programmer can specify what number to set at the start of a random number generation sequence, thereby making the results of the program repeatable and predidictable.$^{33,34,35}$ 
<br><br>
Seed values are set by using *numpy.random.Seed*,$^{10,33,34,35}$ the parameters of which are;

- *a* - Optional; seed value to be set.  If no value is entered, then the default value is none, and the system uses the current time to set the seed to
- *version* - an integer saying how to convert a into an integer, the default is 2

<br>
Another method of seed generation that is important to note is that of *numpy.random.SeedSequence*.$^{33}$ The parameter values for this method include;

- *entropy* - this parameter refers to the ability of the programmer to 'guess' what the seed is$^{36}$
- *spawn_key* - Optional; refers to the number of children created in the first parameter
- *pool_size* - Optional; refers to the pooled entrophy stored in the comptuers internal memory when calling the 'entropy' parameter.
- *n_children_spawned* - Optional; only used if a number of children are already created

<br>
As mentioned above, the mathmateical notion on entropy is an integral part of seed generation.  Entropy is defined by Wolfram Mathworld as being **the amount of disorder in a system**, $^{37}$ which in this instance, would relate to the abiltiy of the programmed to guess what the seed is.  Using Entropy as a basic princible would make guessing the seed for a normal program  that much harder to identify, and predict.
<br><br>
The *SeedSequence* method mixes entropy from different sources to set the inital state for 'independent and very probably not over-lapping BitGenerators'.$^{10}$
<br><br>
For all the security that BitGenerators provide, it is possible for them to be compromised.  Wikipedia outlines the case of the 'Hot Lotto Fraud Scandal'$^{38}$, where an employee compromised the random number generator of the lottery system, to access the seed of the program, and be able to predict the numbers of the lottery draw.
<br><br>
Example 13 below shows a program to roll a dice 3 times.  A seed is set using *np.random.seed()*, where the value of the seed is set to an arbitory number of 6. Running the code multiple times results in the program returning the same result of '9', see Example 13a.

```python
#Example 13 - Roll a 6 sided dice, 3 times

#set the seed of the BitGenerator, to make the results the same. (Ref. 12)
np.random.seed(6)

#roll a six sided dice three times
RollDice= np.random.randint(1,7, size=3)
print(np.sum(RollDice))
```

```python
#Example 13 a - Repeat example 13 to confirm same result

#set the seed of the BitGenerator, to make the results the same. (Ref. 12)
np.random.seed(6)

#roll a six sided dice three times
RollDice= np.random.randint(1,7, size=3)
print(np.sum(RollDice))
```
